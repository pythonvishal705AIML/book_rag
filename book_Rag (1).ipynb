```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
from langchain_community.vectorstores import FAISS
```

```python
from langchain_openai.embeddings import OpenAIEmbeddings
```

```python
import os
```

```python
os.environ["OPENAI_API_KEY"]="put_your_key"
```

```python
mydoc= pdfloader.load()
```

```python
mydoc
```

```python
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
```

```python
text_splitter
```

```python
text = text_splitter.split_documents(mydoc)
```

```python
text
```

```python
embeddings =OpenAIEmbeddings()
```

```python
embeddings
```

```python
db = FAISS.from_documents(text, embeddings)
```

```python
retriever = db.as_retriever()
from longchain.agents_toolkits import create_retriever_tool
tool = create_retriever_tool(retriever, "my book name","my df")

tools =[tool]
```

```python
from langchain.agents.chat_toolkits import create_conversational_retriever_tool
from langchain_openai.chat_models import ChatOpenAI
llm = ChatOpenAI(temprature=0)

agent_executor = create_conversational_retrieval_agent(llm,tools,verbose=True)
```

```python
input=" ask book related que"
result= agent_executor.invoke({"input": input})
```
